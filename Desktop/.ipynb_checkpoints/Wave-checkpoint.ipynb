```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import io

# Constants
mu0 = 4 * np.pi * 10**-7
n = 1000
N = 700  # Number of spatial points
dx = 1.0 * 10**-10  # Spatial resolution in meters
T = 1000 * N  # Number of time steps

# Physical constants
m = 9.10938356e-31  # Electron mass in kg
hbar = 1.0545718e-34  # Reduced Planck's constant in J.s
q = 1.6 * 10**-19  # Elementary charge in C
E = q * 10**-4  # Energy in eV

# Wavenumber and spatial axis
k0 = np.sqrt(2 * m * E) / hbar
X = dx * np.linspace(0, N, N)

# Define Gaussian function
def Gaussian(x, t, sigma):
    return np.exp(-(x - t) ** 2 / (4 * sigma ** 2))

# Initialize a frames array to store each frame
frames = []
frames1 = []
frames2 = []

# Simulation function with frame capturing
def run_simulation(I, color):
    R = 0.01 * (N * dx / np.pi)
    A = (0.5 * mu0 * n * I * R**2) / (N * dx / np.pi)  # Vector potential

    # Potential and wave function setup
    V0 = 0
    sigma = 30.0  # Standard deviation for Gaussian wave packet
    x0 = round(N / 2) - 4 * sigma  # Initial position of the wave packet

    # Precompute constants
    dt = 0.001 * (hbar / (2 * hbar**2 / (m * dx**2)))
    c1 = hbar * dt / (m * dx**2)
    c2 = 2 * dt / hbar

    # Initialize wave function states
    psi_r = np.zeros((3, N))  # Real part
    psi_i = np.zeros((3, N))  # Imaginary part
    PA, PR, FU = 0, 1, 2  # Indices for past, present, future

    # Initial wave packet
    xn = np.arange(1, N)
    x = X[xn] / dx
    gg = Gaussian(x, x0, sigma)
    cx = np.cos(k0 * x)
    sx = np.sin(k0 * x)
    psi_r[PR, xn] = cx * gg
    psi_i[PR, xn] = sx * gg
    psi_r[PA, xn] = cx * gg
    psi_i[PA, xn] = sx * gg

    # Normalize the wave function
    psi_p = psi_r[PR]**2 + psi_i[PR]**2
    P = dx * psi_p.sum()
    nrm = np.sqrt(P)
    psi_r /= nrm
    psi_i /= nrm

    # Time evolution loop
    IDX1 = np.arange(1, N - 1)
    IDX2 = np.arange(2, N)
    IDX3 = np.arange(0, N - 2)

    captured_framesp = []
    captured_framesr = []
    captured_framesi = []
    for t in range(T):  # Adjust time steps for frames
        psi_rPR = psi_r[PR]
        psi_iPR = psi_i[PR]

        # Apply update equations using vectorized operations
        psi_i[FU, IDX1] = psi_i[PA, IDX1] + c1 * (psi_rPR[IDX2] - 2 * psi_rPR[IDX1] + psi_rPR[IDX3]) \
            - (c2 / c1) * q**2 * (A**2) * psi_iPR[IDX1] + (hbar * q / m) * A * (psi_iPR[IDX2] - psi_iPR[IDX3]) / dx
        psi_i[FU] -= c2 * V0 * psi_rPR

        psi_r[FU, IDX1] = psi_r[PA, IDX1] - c1 * (psi_i[PR, IDX2] - 2 * psi_i[PR, IDX1] + psi_i[PR, IDX3]) \
            + (c2 / c1) * q**2 * (A**2) * psi_rPR[IDX1] + (hbar * q / m) * A * (psi_rPR[IDX2] - psi_rPR[IDX3]) / dx
        psi_r[FU] += c2 * V0 * psi_iPR

        # Swap references for next time step
        psi_r[PA], psi_r[PR], psi_r[FU] = psi_rPR, psi_r[FU], psi_r[PA]
        psi_i[PA], psi_i[PR], psi_i[FU] = psi_iPR, psi_i[FU], psi_i[PA]

        # Capture the frame (plot)
        if t % (T//100) == 0:
            psi_p = psi_r[PR]**2 + psi_i[PR]**2
            normalized_psi_p = psi_p / np.sum(psi_p * dx)
            normalized_psi_r = psi_r[PR] / np.sqrt(np.sum(psi_p * dx))
            normalized_psi_i = psi_i[PR] / np.sqrt(np.sum(psi_p * dx))
            captured_framesp.append(normalized_psi_p)
            captured_framesr.append(normalized_psi_r)
            captured_framesi.append(normalized_psi_i) 
        t +=dt    

    return captured_framesp, captured_framesr, captured_framesi
# Capture both simulations
psi_p_no_A_frames, psi_r_no_A_frames, psi_i_no_A_frames = run_simulation(I=0, color='b')  # No vector potential
psi_p_with_A_frames, psi_r_with_A_frames, psi_i_with_A_frames = run_simulation(I=70/(dx**2), color='r') # With large vector potential
psi_p_with_AA_frames, psi_r_with_AA_frames, psi_i_with_AA_frames = run_simulation(I=-70/(dx**2), color='orange')

# Combine and save frames as GIF without saving PNGs
for i in range(len(psi_p_no_A_frames)):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(X, psi_p_no_A_frames[i], label="No Vector Potential", color='b')
```

```python
ax.plot(X, psi_p_with_A_frames[i], label="With +Vector Potential", color='r', alpha = 0.5)
    ax.plot(X, psi_p_with_AA_frames[i], label="With -Vector Potential", color='orange', alpha = 0.5)
    ax.set_xlim(1e-8, X.max())
    ax.set_ylim(0, 15*10**7)
    ax.set_xlabel("Position (m)")
    ax.set_ylabel("Probability Density")
    ax.set_title(f"Wave Function at t = {i}")
    ax.legend()
    ax.grid(True)

    # Save each frame to a BytesIO object
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    
    # Convert the buffer into an image
    frames.append(Image.open(buf))
    plt.close()

# Create and save the GIF in memory
frames[0].save("wavefunction_evolution.gif", save_all=True, append_images=frames[1:], duration=100, loop=0)

print("GIF created and saved as 'wavefunction_evolution.gif'")


for i in range(1, len(psi_r_with_A_frames)):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(X, psi_r_no_A_frames[i], label="No Vector Potential", color='b')
    ax.plot(X, psi_r_with_A_frames[i], label="With +Vector Potential", color='r', alpha = 0.5)
    ax.plot(X, psi_r_with_AA_frames[i], label="With -Vector Potential", color='orange', alpha = 0.5)
    ax.set_xlim(0, X.max())
    ax.set_ylim(-12000, 12000)
    ax.set_xlabel("Position (m)")
    ax.set_ylabel("wave function")
    ax.set_title(f"Wave Function at t = {i}")
    ax.legend()
    ax.grid(True)

    # Save each frame to a BytesIO object
    buf1 = io.BytesIO()
    plt.savefig(buf1, format='png')
    buf1.seek(0)
    
    # Convert the buffer into an image
    frames1.append(Image.open(buf1))
    plt.close()

# Create and save the GIF in memory
frames1[0].save("real_wavefunction_evolution.gif", save_all=True, append_images=frames1[1:], duration=100, loop=0)

print("GIF created and saved as 'wavefunction_evolution.gif'")
```
